```python
student_name = input("Enter your name")
print("Enter your skills score please 1-10")

programming= int(input("Programming: "))
desing= int(input("Desing: "))
networking= int(input("Networking: "))
data_analysis = int(float(input("Data Analysis: ")))

student = {
    "name": student_name,
    "skills scores": {
        "programming": programming,
        "desing": desing,
        "networking": networking,
        "data_analysis": data_analysis
    }
}

print("Your skills scores are:")
print(f"name: {student['name']}")
print(f"programming: {student['skills scores']['programming']}")
print(f"desing: {student['skills scores']['desing']}")
print(f"networking: {student['skills scores']['networking']}")
print(f"data_analysis: {student['skills scores']['data_analysis']}")


highest_score = max(student["skills scores"]["programming"], student["skills scores"]["desing"], student["skills scores"]["networking"], student["skills scores"]["data_analysis"])

highest_count = 0
if student["skills scores"]["programming"] == highest_score:
    highest_count += 1
if student["skills scores"]["desing"] == highest_score:
    highest_count += 1
if student["skills scores"]["networking"] == highest_score:
    highest_count += 1
if student["skills scores"]["data_analysis"] == highest_score:
    highest_count += 1

if highest_count > 1:
    print("Your recommended Career is multiple careers paths identifyed")
elif highest_score == student["skills scores"]["programming"]:
    print("Your recommended Career is programming")
elif highest_score == student["skills scores"]["desing"]:
    print("Your recommended Career is desing")    
elif highest_score == student["skills scores"]["networking"]:
    print("Your recommended Career is networking")
elif highest_score == student["skills scores"]["data_analysis"]:
    print("Your recommended Career is data analysis")
```
