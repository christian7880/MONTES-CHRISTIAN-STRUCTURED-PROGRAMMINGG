A recursive function is a function that calls itself.

Observe the following example.



````python
def countdown(n):

    print(n)

    if n > 1:
        countdown(n - 1)

countdown(5)
````



Execute the program.

### Questions

1. Which numbers are displayed?

--- the sequence of 5 to 1

2. What happens when `n` becomes 1?

--- It prints 1, and since the condition n > 1 is false, it stops making recursive calls

3. Why does the program stop?

--- Because the base case is reached


# Part 7. Understanding the Base Case

Every recursive function needs a condition that stops the recursion.

Observe the following example.



````python
def say_hello(n):

    if n == 0:
        return

    print("Hello")
    say_hello(n - 1)

say_hello(4)
````

### Questions

1. How many times is "Hello" printed?

--- Four times

2. What is the base case?

--- if n == 0

3. What would happen if the `if` statement were removed?

--- the program crashed

# Part 9. First Recursive Mathematical Example

The factorial of a number is defined as:

```text
4! = 4 × 3 × 2 × 1
```

Observe the following code.

```python
def factorial(n):

    if n == 1:
        return 1

    return n * factorial(n - 1)

print(factorial(4))
```

1. What is the difference between a function and a procedure?

---
A function executes a block of code and returns a specific value using a return statement. A procedure executes a sequence of instructions but does not return any value; it usually just performs an action like printing to the console
---

2. What is recursion?

---
Recursion is a programming technique where a function solves a problem by calling itself, breaking down the main task into smaller and simpler instances of the same problem.
---

3. Why does every recursive function need a base case?

---
Because without a base case, the function would keep calling itself infinitely, eventually running out of memory and causing a 'Stack Overflow' error. The base case provides the stopping condition
---

4. Which topic was most difficult for you?

````python
# Final challenge

# --- PROCEDURES ---
def display_menu():
    print("\n=======================")
    print("      MAIN MENU        ")
    print("=======================")
    print("1. Calculate Area (Rectangle)")
    print("2. Calculate Average")
    print("3. Countdown")
    print("4. Exit")
    print("=======================")

def say_goodbye():
    print("\nThank you for using the program. Goodbye!")


def calculate_area(base, height):
    return base * height

def calculate_average(num1, num2, num3):
    return (num1 + num2 + num3) / 3


def countdown(n):
    print(n)
    if n > 1:
        countdown(n - 1)


def main():
    while True:
        display_menu()
        try:
            option = int(input("Select an option (1-4): "))

            if option == 1:
                b = float(input("Enter the base of the rectangle: "))
                h = float(input("Enter the height of the rectangle: "))
                area = calculate_area(b, h)
                print(f"-> The area of the rectangle is: {area}")

            elif option == 2:
                n1 = float(input("Enter the first number: "))
                n2 = float(input("Enter the second number: "))
                n3 = float(input("Enter the third number: "))
                avg = calculate_average(n1, n2, n3)
                print(f"-> The average is: {avg}")

            elif option == 3:
                start = int(input("Enter a starting number for the countdown: "))
                print("Starting countdown:")
                countdown(start)

            elif option == 4:
                say_goodbye()
                break
            else:
                print("Invalid option. Please choose a number between 1 and 4.")

        except ValueError:
            print("Please enter a valid numeric value.")

# Run the program
main()
````
